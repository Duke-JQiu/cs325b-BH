```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array of x values from 0 to 2 * pi (one full sine wave cycle)
x = np.linspace(0, 2 * np.pi, 1000)

# Compute the sine value for each x value
y = np.sin(x)

# Plot the sine curve
plt.plot(x, y)

# Provide title and labels
plt.title("Sine Curve")
plt.xlabel("x values")
plt.ylabel("y = sin(x)")

# Display the plot
plt.grid(True)
plt.show()
```
